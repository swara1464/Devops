# ML Project Environment Setup Guide

This notebook documents the steps to set up a simple Machine Learning project environment.

## Step 1: Create requirements.txt

Define the necessary packages in a `requirements.txt` file.

```text
numpy
pandas
scikit-learn
matplotlib
jupyter
```

## Step 2: Install Packages

Run the following command to install the dependencies:

```bash
pip install -r requirements.txt
```

````python
import numpy as np
import pandas as pd
import sklearn

print("Numpy version:", np.__version__)
print("Pandas version:", pd.__version__)
print("Scikit-learn version:", sklearn.__version__)
print("\nSuccess! Environment is set up correctly.")
````

````python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

print("numpy:", np.__version__)
print("pandas:", pd.__version__)

# tiny sample
X = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])

model = LinearRegression().fit(X, y)
print("coef:", model.coef_, "intercept:", model.intercept_)
````
